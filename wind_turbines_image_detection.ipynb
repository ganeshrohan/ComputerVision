```python
#Import YOLO 
from ultralytics import YOLO
from IPython.display import display, Image
```

```python
# load a pretrained model (recommended for training)
model = YOLO("yolov8n.pt")
```

```python
model.info()
```

### Model Training

```python
# Train the model on the Pushpin example dataset for 300 epochs
# Default early stopping with patience=100

results = model.train(
    data="/kaggle/input/windturbines/data.yaml",
    epochs=25,
    imgsz=640,
    save=True,
    device=0,
    plots=True,
    batch=-1,
    project='/kaggle/working/windturbines',
    name='yolov8_25'
  )
```

### Evaluation

```python
validation_results = model.val(
    data="/kaggle/input/windturbines/data.yaml",
    imgsz=640,
    batch=16,
    conf=0.25,
    iou=0.5,
    device="0"
  )
```

```python
print(validation_results.box.map50)  # mAP50
print(validation_results.box.map75)  # mAP75
```

```python
Image(filename=f'/kaggle/working/windturbines/yolov8_25/results.png', width=600)
```

### Model Inference

```python
# Run batched inference on a list of images
results = model(
    source='/kaggle/input/windturbines/test/images/DJI_0010_03_05_png.rf.3c27c0c4c47071574c6a27d83d992dba.jpg',
    iou=0.5,
    conf=0.25
)  # return a generator of Results objects

from IPython import display
display.clear_output()


# Process results generator
for result in results:
    boxes = result.boxes  # Boxes object for bounding box outputs
    detected_classes = [box.cls for box in boxes]  # Get the detected classes
    classes = [int(cls.item()) for cls in detected_classes]
    result.show()
```
